```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
```

```python
img = imread('man1.jpg')
```

```python
imshow(img)
```

```python
from sklearn.cluster import KMeans
```

```python
img.shape
```

```python
X = img.reshape(-1,4)/255
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
X.shape
```

```python
kmeans.fit(X)
```

```python
idenclusters = kmeans.fit_predict(X)
```

```python
idenclusters 
```

```python
kmeans.labels_
```

```python
new_X = kmeans.cluster_centers_[kmeans.labels_]
```

```python
image = new_X.reshape(168, 300, 3)
```

```python
imshow(image)
```
